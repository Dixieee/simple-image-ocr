```python
from google.colab.patches import cv2_imshow
import cv2
import pytesseract

# Load image
image = cv2.imread('/content/contoh documentation.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Langsung OCR ke seluruh gambar
print("=== OCR langsung tanpa kontur ===")
text = pytesseract.image_to_string(gray, config='--psm 6')
print(text.strip())

cv2_imshow(image)
```

```python
with open("/content/hasil_ocr.txt", "w", encoding="utf-8") as f:
    f.write(text)
```
